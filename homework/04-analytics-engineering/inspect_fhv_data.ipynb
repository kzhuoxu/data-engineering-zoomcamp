```python
FILE_NAME = 'fhv_tripdata_2019-02'
BASE_URL = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/fhv/'
```

```python
# Download data from url
import urllib.request
urllib.request.urlretrieve(BASE_URL + FILE_NAME + '.csv.gz', FILE_NAME + '.csv.gz')
```

```python
import gzip
import shutil

with gzip.open(FILE_NAME + '.csv.gz', 'rb') as f_in:
    with open(FILE_NAME + '.csv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```

```python
import pandas as pd
fhv_data = pd.read_csv(FILE_NAME + '.csv')
```
